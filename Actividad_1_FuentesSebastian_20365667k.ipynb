### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.


R: 
El ruido rosa es un tipo de sonido que tiene una distribución de frecuencia en la que la potencia disminuye en aproximadamente 3 decibeles por octava a medida que la frecuencia aumenta. Esto significa que las frecuencias más bajas tienen más energía que las frecuencias más altas. El ruido rosa es comúnmente utilizado en aplicaciones de audio y acústica, ya que tiene una distribución de energía que se asemeja a la forma en que los humanos perciben el sonido.

Por otro lado, el ruido gaussiano, también conocido como ruido blanco, es un tipo de señal aleatoria en la que cada muestra tiene una distribución de probabilidad gaussiana, lo que significa que las amplitudes de las muestras están distribuidas de acuerdo con una campana de Gauss. En otras palabras, el ruido gaussiano tiene una distribución de amplitudes donde cada valor es igualmente probable.

La diferencia principal entre el ruido rosa y el ruido gaussiano radica en cómo se distribuye la energía en función de la frecuencia. Mientras que el ruido rosa tiene más energía en las frecuencias bajas y menos en las frecuencias altas, el ruido gaussiano tiene energía distribuida de manera uniforme en todas las frecuencias.






```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np

def generate_pink_noise(length):
    # Generar ruido blanco
    white_noise = np.random.randn(length)
    
    # Aplicar el filtro 1/f
    pink_noise = np.cumsum(white_noise) / np.sqrt(np.arange(1, length+1))
    
    return pink_noise

length = 10000 # Definir la longitud del ruido
samplerate = 44800

# Generar ruido rosa
pink_noise = generate_pink_noise(length)

# Generar ruido gaussiano
gaussian_noise = np.random.normal(size=length)

# Calcular la FFT de los ruidos
fft_pink = np.fft.fft(pink_noise)
fft_gaussian = np.fft.fft(gaussian_noise)

# Calcular la densidad espectral de potencia (PSD)
psd_pink = np.abs(fft_pink) ** 2
psd_gaussian = np.abs(fft_gaussian) ** 2

# Calcular la frecuencia correspondiente a cada bin de la FFT
freq = np.fft.fftfreq(length, 1/samplerate)

# Convertir la PSD a dB
psd_pink_db = 10 * np.log10(psd_pink)
psd_gaussian_db = 10 * np.log10(psd_gaussian)

# Visualizar los ruidos
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(freq[freq > 0], psd_pink_db[freq > 0], color='pink', label='Ruido Rosa')
plt.title('Ruido Rosa')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Intensidad (dB)')
plt.legend()
plt.subplot(1,2,2)
plt.plot(freq[freq > 0], psd_gaussian_db[freq > 0], color='blue', label='Ruido Gaussiano')
plt.title('Ruido Gaussiano')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Intensidad (dB)')
plt.legend()
plt.show()


```
